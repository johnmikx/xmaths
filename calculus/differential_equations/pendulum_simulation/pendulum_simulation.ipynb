# **Pendulum Motion Simulation**

Simulates the motion of a pendulum by solving a second-order ordinary differential equation (ODE) using the Euler method. The simulation visualizes the angle of the pendulum over time. This approach is inspired by the concepts discussed in **3Blue1Brown's** "[Differential equations, a tourist's guide | DE1](https://www.youtube.com/watch?v=p_di4Zn4wz4&list=PLZHQObOWTQDNPOjrT6KVlfJuKtYTftqH6)" video series.

```python
import numpy as np
```

```python
# Physical Constants
g = 9.8
L = 2
mu = 0.1

THETA_0 = np.pi / 3 # 60 degrees
THETA_DOT_0 = 0 # no angular velocity

# Definition to the ODE
def get_theta_double_dot(theta, theta_dot):
  return -mu * theta_dot - (g/L) * np.sin(theta)

# Solution to the Differential Equation
def theta(t):
  # Initialize Changing Values
  theta = THETA_0
  theta_dot = THETA_DOT_0
  delta_t = 0.01 # some time step
  for time in np.arange(0, t, delta_t):
    theta_double_dot = get_theta_double_dot(
        theta, theta_dot
        )
    theta += theta_dot * delta_t
    theta_dot += theta_double_dot * delta_t
  return theta
```

### **Task**
Simulate a pendulum's motion using the `theta` function and visualize the results.

## **Generate Time Values**

Create a range of time values for the simulation.


**Reasoning**:
Generate a NumPy array of time values from 0 to 10 with a step of 0.01.



```python
time_values = np.arange(0, 10.01, 0.01)
```

## **Calculate Theta Values**

Apply the `theta` function to my time values to get the corresponding pendulum angles.


**Reasoning**:
Apply the `theta` function to each time value and store the results in a new array.



```python
theta_values = np.array([theta(t) for t in time_values])
```

## **Visualize the Results**

Visualize the results by plotting the calculated theta values against the time values.


**Reasoning**:
Visualize the calculated theta values against the time values by creating a line plot with appropriate labels, title, and grid.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(time_values, theta_values)
plt.title("Pendulum Motion Simulation", fontweight='bold')
plt.xlabel("Time (s)", fontweight='bold')
plt.ylabel("Angle (radians)", fontweight='bold')
plt.grid(True)
plt.show()
```
