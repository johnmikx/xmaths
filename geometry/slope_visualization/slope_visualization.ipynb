```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define Points (Example)
x1, y1 = 0, 0
x2, y2 = 2, 2

# Create x Range for the Line
x = np.linspace(x1 - 1, x2 + 1, 100)
y = (y2 - y1)/(x2 - x1) * (x - x1) + y1 # Line Equation

# Plot the Line
plt.plot(x, y, 'b-', label='Line $(L)$')
plt.plot(x1, y1, 'bo', label=f'$P1({x1},{y1})$')
plt.plot(x2, y2, 'bo', label=f'$P2({x2},{y2})$')

# Add Rise and Run Annotations
plt.plot([x1, x2], [y1, y2], 'r-', linewidth=2) # Run
plt.plot([x2, x2], [y1, y2], 'r-', linewidth=2) # Rise
plt.text((x1 + x2) / 2, y1 - 0.2, f'$\Delta x = {x2 - x1} (run)$', ha='center')
plt.text(x2 + 0.1, (y1 + y2) / 2, f'$\Delta y = {y2 - y1} (rise)$', va='center')

# Add Labels and Title
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.5) # x-axis
plt.axvline(x=0, color='k', linestyle='-', linewidth=0.5) # y-axis
plt.xlabel('x', fontweight='bold')
plt.ylabel('y', fontweight='bold')
plt.title('Slope of Line $(L)$', fontweight='bold')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
