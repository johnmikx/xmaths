```python
import matplotlib.pyplot as plt
import math
```

```python
def plot_distance(x1, y1, x2, y2):
  # Coordinates
  x3, y3 = x2, y1 # Right-angle Corner

  # Plot Points
  plt.figure(figsize=(6,5))
  plt.plot([x1, x2], [y1, y2], 'r-', label=r'$|P_1P_2|$')
  plt.plot([x1, x3], [y1, y3], 'b-', label=r'$|P_1P_3|$')
  plt.plot([x2, x3], [y2, y3], 'b-', label=r'$|P_2P_3|$')

  # Scatter Points
  plt.scatter([x1, x2, x3], [y1, y2, y3], color='black')
  plt.text(x1, y1, f'P1({x1},{y1})', fontsize=9, ha='right')
  plt.text(x2, y2, f'P2({x2},{y2})', fontsize=9, ha='left')
  plt.text(x3, y3, f'P3({x3},{y3})', fontsize=9, ha='center', va='top')

  # Axes and grid
  plt.axhline(0, color='gray', linewidth=1)
  plt.axvline(0, color='gray', linewidth=1)
  plt.grid(True, linestyle='--', alpha=0.6)
  plt.xlabel('x', fontweight='bold')
  plt.ylabel('y', fontweight='bold')
  plt.title('Distance Between Two Points with Right Triangle', fontweight='bold')
  plt.axis('equal')
  plt.legend()
  plt.tight_layout()
  plt.show()
```

```python
def distance(x1, y1, x2, y2):
  return math.sqrt((x2 - x2)**2 + (y2 - y1)**2)
```

```python
# Example 1:

x1, y1 = 1, 2
x2, y2 = 4, 6
d = distance(x1, y1, x2, y2)
print(f'The distance between P1({x1},{y1}) and P2({x2},{y2}) is {d:.2f}')
plot_distance(x1, y1, x2, y2)
```

```python
# Example 2:

x1, y1 = 1, 1
x2, y2 = 3, 3
d = distance(x1, y1, x2, y2)
print(f'The distance between P1({x1},{y1}) and P2({x2},{y2}) is {d:.2f}')
plot_distance(x1, y1, x2, y2)
```

```python
# Example 3:

x1, y1 = 0, 1
x2, y2 = 4, 4
d = distance(x1, y1, x2, y2)
print(f'The distance between P1({x1},{y1}) and P2({x2},{y2}) is {d:.2f}')
plot_distance(x1, y1, x2, y2)
```
