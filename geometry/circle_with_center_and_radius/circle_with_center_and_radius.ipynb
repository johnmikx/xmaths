```python
import matplotlib.pyplot as plt
import math
```

```python
def plot_circle_with_radius(h, k, x, y):
  # Calculate Radius
  r = math.sqrt((x - h)**2 + (y - k)**2)

  # Circle Points
  circle = plt.Circle((h, k), r, color='red', fill=False)

  # Plot Setup
  fig, ax = plt.subplots(figsize=(6,6))
  ax.add_patch(circle)
  ax.plot([h, x], [k, y], 'b-', label='Radius')
  ax.scatter([h, x], [k, y], color='blue')

  # Label Points
  ax.text(h, k, f'C({h},{k})', fontsize=10, ha='right')
  ax.text(x, y, f'P({x},{y})', fontsize=10, ha='left')

  # Radius Label (Midpoint of Line from Center C to Point P)
  mid_x = (h + x) / 2
  mid_y = (k + y) / 2
  ax.text(mid_x, mid_y, f'r = {r:.2f}', fontsize=9, color='blue', ha='right', va='center')

  # Axes and Grid
  ax.axhline(0, color='gray', linewidth=1)
  ax.axvline(0, color='gray', linewidth=1)
  ax.set_xlabel('x', fontweight='bold')
  ax.set_ylabel('y', fontweight='bold')
  ax.set_title('Circle with Center (C) and Radius to Point (P)', fontweight='bold')
  ax.grid(True, linestyle='--', alpha=0.6)
  ax.legend()
  plt.tight_layout()
  plt.show()
```

```python
# Example 1: Center (0, 0) and Point (1, 1)
h, k = 0, 0
x, y = 1, 1
plot_circle_with_radius(h, k, x, y)
```

```python
# Example 2: Center (2, 3) and Point (5, 7)
h, k = 2, 3
x, y = 5, 7
plot_circle_with_radius(h, k, x, y)
```
